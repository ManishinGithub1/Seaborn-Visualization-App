```python
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
```

```python
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
sns.set(style="darkgrid")
```

```python
tips.to_csv("tips.csv",index=False)
```

```python
import os
os.getcwd()
```

```python
#Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",size="size",palette="deep")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
#Line Plot
sns.lineplot(data=tips,x="size",y="total_bill",hue='sex',ci=None,marker='o')
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
#Bar Plot
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',palette='muted')
plt.title("Barplot of Total Bill by Day")
plt.show()
```

```python
#Box Plot
sns.boxplot(data=tips,x='day',y='tip',hue='smoker',palette='Set2')
plt.title("Boxplot of Tips by Day")
plt.show()
```

```python
#Violin Plot
sns.violinplot(data=tips,x='day',y='total_bill',hue='time',split=True,palette='bright')
plt.title("Violinplot of Total bill by Day")
plt.show()
```

```python
#Count Plot
sns.countplot(data=tips,x='day',hue='smoker',palette='dark')
plt.title("Countplot of Days by Smoker Status")
plt.show()
```

```python
#Regression Plot
sns.regplot(data=tips,x='total_bill',y='tip',scatter_kws={'s':50},line_kws={'color':'red'})
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

```python
#Histogram with KDE
sns.histplot(data=tips,x='total_bill',bins=15,kde=True,color='blue')
plt.title("Histogram of Total Bill")
plt.show()
```

```python
#Pair Plot
sns.pairplot(tips,hue='sex',vars=['total_bill','tip','size'],palette='husl')
plt.suptitle("Pair Plot: numeric Variables by Gender",y=1.02)
plt.show()
```

```python
#Cat Plot
sns.catplot(data=tips,x='day',y='tip',hue='sex',kind='point',palette='bright')
plt.title("Catplot(Points): Tips by Day and Gender")
plt.show()
```

```python
#Joint Plot

sns.jointplot(data=tips,x="total_bill",y='tip',kind='scatter',color='purple',hue='sex',palette='pastel')
plt.suptitle("Joint Plot: Total Bill vs Tip",y=1.02)
plt.show()
```

```python
#Facet Grid
g=sns.FacetGrid(tips,col='time',row='smoker',margin_titles=True).map(sns.histplot,'total_bill',bins=10,kde=True).add_legend()
g
```

```python
#Strip Plot
sns.stripplot(data=tips,x='day',y='tip',hue='sex',jitter=True,palette='Set1')
plt.title("Strip Plot: Tips by data and gender")
plt.show()
```

```python
#Kde Plot
sns.kdeplot(data=tips,x='total_bill',y='tip',fill=True,hue='sex',palette='tab10')
plt.title("KDE Plot: Total Bill vs Tip")
plt.show()
```
